```python
import pandas as pd
import plotly.express as px    #data visualization library which makes easy and quick plots
import plotly.graph_objects as go  # makes advanced and customizable graph
import plotly.io as pio   #to customize graph templates
import plotly.colors as colors
pio.templates.default="plotly_white"  #white theme by default for graph
```

```python
data=pd.read_csv("Sample - Superstore.csv", encoding="latin-1")  # to read special characters, we used encoding
data.head()
```

# Descriptive Statistics of the dataset

```python
data.describe()
```

```python
data.info()
```

# Converting Order Date and Ship Date columns to datetime

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

# Adding new columns

```python
data['Order Month']=data['Order Date'].dt.month
data['Order Year']=data['Order Date'].dt.year
data['Order Day of Week']=data['Order Date'].dt.dayofweek
```

```python
data.head()
```

# 1. Calculate Monthly sales of store and identify which month had the highest sales and which month had the lowest sales

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales_by_month, x='Order Month', y='Sales', title='Monthly Sales Analysis')
fig.show()
```

# 2. Analyze Sales based on product categories and determine which category has the lowest sales and highest sales

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
fig = px.pie(sales_by_category, values='Sales', names='Category', hole=0.5, color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category', title_font=dict(size=24))
fig.show()
```

# 3. Sales Analysis by Sub-Category

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
fig = px.bar(sales_by_subcategory, x='Sub-Category', y='Sales', title='Sales Analysis by Sub-Category')
fig.show()
```

# 4. Analyze Monthly profit from sales and determine which month had the highest profit

```python
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()
fig = px.line(profit_by_month, x='Order Month', y='Profit', title='Monthly Profit Analysis')
fig.show()
```

# 5. Analyze the profit by category and sub-category

```python
data.head()
```

```python
profit_by_category=data.groupby('Category')['Profit'].sum().reset_index()
fig=px.pie(profit_by_category, values='Profit', names='Category', hole=0.5, color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Profit Analysis by Category', title_font=dict(size=24))
fig.show()
```

```python
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_subcategory, x='Sub-Category', y='Profit', title='Profit Analysis by Sub-Category')
fig.show()
```

# Analyze the Sales and Profit by Customer Segmentation

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
color_palette = colors.qualitative.Pastel
fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'], y=sales_profit_by_segment['Sales'], name='Sales', marker_color=color_palette[0]))
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'], y=sales_profit_by_segment['Profit'], name='Profit', marker_color=color_palette[1]))
fig.update_layout(title='Sales and Profit Analysis by Customer Segment', xaxis_title='Customer Segment', yaxis_title='Amount')
fig.show()
```

# Analyse the sales to profit ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
